# Matriz de Correlaci贸n

### Cargamos las librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Cargamos nuestro dataset

```python
houses = pd.read_csv('./data/Melbourne_housing.csv')
```

Creamos nuestra matriz de correlaci贸n

```python
matrix = houses.corr()
```

Dado que nuestra matriz de correlaci贸n es simetrica podemos creamos una matriz triangular para quitar datos que pueden terminar siendo redundantes. Cambiamos a un formato booleano

```python
mask = np.triu(np.ones_like(matrix, dtype=bool))
```

Creamos una escala de color para nuestra visualizaci贸n

```python
cmap = sns.diverging_palette(250, 15, s=75, l=40, n=9, center='light', as_cmap=True)
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(matrix, mask=mask, center=0, annot=True, fmt='.2f', square=True, cmap=cmap)
plt.show()
```

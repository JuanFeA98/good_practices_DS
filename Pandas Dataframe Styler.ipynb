# Pandas DataFrame Styler

## Cargamos nuestras librerias

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

## Cargamos los datos

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.describe()
```

### Creamos una tabla cruzada

```python
pd.crosstab(diamonds.cut, diamonds.clarity)
```

### Agregamos estilos

#### Análisis por columnas

```python
pd.crosstab(diamonds.cut, diamonds.clarity).\
        style.background_gradient(cmap='PuBu')
```

#### Análisis por registros

```python
pd.crosstab(diamonds.cut, diamonds.clarity).\
        style.background_gradient(cmap='PuBu', axis=1)
```

#### Hacemos el análisis sobre el precio promedio de los diamantes con respecto a su corte y su claridad

```python
agg_prices = pd.crosstab(diamonds.cut, diamonds.clarity,
                aggfunc=np.mean, values=diamonds.price).\
                style.background_gradient(cmap='PuBu')

agg_prices.format('{:.2f}')
```

# **Básicos de Matplotlib**

## **Librerías**

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import ptitprince as pt
```

```python
import warnings

warnings.filterwarnings('ignore')
```

## **Datos**

```python
iris = sns.load_dataset('iris')
```

```python
iris.sample()
```

### **Diagramas de dispersión** 

#### **Diagrama de nubes** 

```python
fig, ax = plt.subplots(figsize=(10, 7))

ax = pt.half_violinplot(
    data=iris,
    x='sepal_length',
    y='species',
    width=0.5    
)

plt.show()
```

#### **Diagrama de lluvia** 

```python
fig, ax = plt.subplots(figsize=(10, 7))

ax = pt.half_violinplot(
    data=iris,
    x='sepal_length',
    y='species',
    width=0.5    
)

ax = sns.stripplot(
    data=iris,
    x='sepal_length',
    y='species',
    jitter=1
)

plt.show()
```

#### **Diagrama de lluvia con diagrama de caja** 

```python
fig, ax = plt.subplots(figsize=(10, 7))

ax = pt.half_violinplot(
    data=iris,
    y='species',
    x='sepal_length',
    width=0.5    
)

ax = sns.stripplot(
    data=iris,
    y='species',
    x='sepal_length',
    jitter=1
)

ax = sns.boxplot(
    data=iris,
    y='species',
    x='sepal_length',
    width=0.15
)

plt.show()
```

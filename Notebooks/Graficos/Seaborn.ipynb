## **Librerías**

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
```

```python
warnings.filterwarnings('ignore')
```

## **Datos**

```python
tips = sns.load_dataset('tips')
```

```python
tips.sample()
```

## **Gráficos**

### **Datos Categóricos**

#### **Diagrama de recuento**

```python
fig, ax = plt.subplots(figsize=(12, 3))

sns.countplot(
    data=tips,
    x='day'
)

plt.title('Conteo de propinas por días de la semana')

plt.show()
```

#### **Diagrama de recuento Múltiple**

```python
fig, ax = plt.subplots(figsize=(12, 3))

sns.countplot(
    data=tips,
    x='day',
    hue='sex'
)

plt.title('Conteo de propinas por días de la semana')

plt.show()
```

#### **Gráfico de Barras**

```python
fig, ax = plt.subplots(figsize=(12, 3))

sns.barplot(
    data=tips,
    x='day',
    y='tip',
    ci=None
)

plt.title('Valor promedio de las propinas por días de la semana')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 3))

sns.barplot(
    data=tips,
    x='day',
    y='tip',
    estimator=np.sum,
    ci=None
)

plt.title('Valor total de las propinas por días de la semana')

plt.show()
```

#### **Gráfico de Barras Múltiple**

```python
fig, ax = plt.subplots(figsize=(12, 3))

sns.barplot(
    data=tips,
    x='day',
    y='tip',
    estimator=np.sum,
    hue='sex',
    ci=None
)

plt.title('Total de las propinas por los días de la semana y Genero')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 3))

sns.barplot(
    data=tips,
    x='day',
    y='tip',
    estimator=np.mean,
    hue='sex',
    ci=None
)

plt.title('Valor promedio de las propinas por días de la semana y Genero')

plt.show()
```

### **Datos Categóricos y Numéricos**

#### **Gráfico de Enjambre**

```python
fig, ax = plt.subplots(figsize=(12, 3))

sns.swarmplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='day',
    legend=False
)

plt.title('Distribución de la cuenta total por día')
plt.xlabel('Día de la semana')
plt.ylabel('Cuenta Total')

plt.show()
```

#### **Diagrama de Caja**

```python
fig, ax = plt.subplots(figsize=(12, 3))

sns.boxplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='day'
)

plt.title('Distribución de la cuenta total por día')
plt.xlabel('Día de la semana')
plt.ylabel('Cuenta Total')

plt.show()
```

### **Datos Numéricos**

#### **Gráfico de Densidad**

```python
fig, ax = plt.subplots(figsize=(12, 3))

sns.kdeplot(
    data=tips['total_bill'],
    fill=True,
    palette='muted'
)

plt.title('Densidad de la cuenta total')
plt.xlabel('Cuenta Total')
plt.ylabel('Densidad')
plt.show()
```

#### **Gráfico de Distribución**

```python
sns.displot(
    data=tips,
    x='total_bill',
    kde=True
)

plt.title('Densidad de la cuenta total')
plt.xlabel('Cuenta Total')
plt.ylabel('Densidad')

plt.show()
```

#### **Gráfico de Dispersión**

```python
fig, ax = plt.subplots(figsize=(12, 3))

sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex'
)

plt.title('Relación entre la cuenta total y las propinas')
plt.xlabel('Cuenta Total')
plt.ylabel('Propinas')
plt.show()
```

#### **Mapa de calor**

```python
correlation_matrix = tips.corr(numeric_only=True)
correlation_matrix

fig, ax = plt.subplots(figsize=(5, 3.5))

sns.heatmap(
    data=correlation_matrix,
    annot=True
)

plt.title('Matriz de Correlación')
plt.show()
```

#### **Diagrama de correlación**

```python
sns.pairplot(
    data=tips,
    hue='day'
)

plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
import sweetviz as sv
```

## **Datos**

```python
df = pd.read_csv('../../data/Melbourne_housing.csv')
```

```python
df.dropna(subset='Price', inplace=True)
```

```python
X = df.drop('Price', axis=1)
y = df[['Price']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=13)
```

```python
df_train = pd.merge(X_train.reset_index(), y_train.reset_index()).drop('index', axis=1)
df_test = pd.merge(X_test.reset_index(), y_test.reset_index()).drop('index', axis=1)
```

## **Reporte**

```python
profile = sv.compare(
    source=df_train,
    compare=df_test,
    target_feat='Price'
)
```

```python
profile.show_notebook()
```
